```python
#importing matplot library
import matplotlib.pyplot as plt
```

```python
#importing numpy
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x,y,z)
```

```python
#sine wave plot
plt.plot(x,y)
plt.show()
```

```python
#adding labels and title
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("sine wave")
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("parabolic")
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g*')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
Languages=["English","Hindi","German","French"]
People=[100,50,200,75]
ax.bar(Languages,People)
ax.set_xlabel("Languages")
ax.set_ylabel("No of speakers")
ax.set_title("Bar Chart")
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
Languages=["English","Hindi","German","French"]
People=[100,50,200,75]
explode=(0,0.1,0,0)
ax.pie(People,explode=explode,labels=Languages,autopct='%.1f%%',startangle=90)
plt.show()
```

```python
x=np.linspace(1,10,30)
y=np.cos(x)
z=np.sin(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
